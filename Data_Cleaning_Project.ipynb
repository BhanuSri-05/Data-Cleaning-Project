```python
import pandas as pd
import numpy as np

# ============================
# STEP 1: LOAD DATASET
# ============================

df = pd.read_csv("Sample - Superstore.csv", encoding="cp1252")

print("Dataset Shape:")
print(df.shape)

print("\nFirst 5 Rows:")
print(df.head())

# ============================
# STEP 2: CHECK MISSING VALUES
# ============================

print("\nMissing Values:")
print(df.isnull().sum())

# ============================
# STEP 3: HANDLE MISSING VALUES
# ============================

# Fill numeric columns with median

numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Fill text columns with 'Unknown'

text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col].fillna("Unknown", inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# ============================
# STEP 4: CHECK DUPLICATES
# ============================

duplicate_count = df.duplicated().sum()

print("\nDuplicate Rows Found:", duplicate_count)

# Remove duplicates

df.drop_duplicates(inplace=True)

print("Duplicates Removed Successfully")

# ============================
# STEP 5: CLEAN TEXT COLUMNS
# ============================

for col in text_cols:

    # Remove extra spaces
    df[col] = df[col].astype(str).str.strip()

    # Convert to title case
    df[col] = df[col].str.title()

print("\nText Formatting Cleaned")

# ============================
# STEP 6: STANDARDIZE DATE COLUMNS
# ============================

# Replace with actual date column names

date_columns = []

for col in df.columns:

    if "date" in col.lower():
        date_columns.append(col)

for col in date_columns:

    try:
        df[col] = pd.to_datetime(
            df[col],
            errors='coerce'
        )

        df[col] = df[col].dt.strftime('%Y-%m-%d')

    except:
        pass

print("\nDate Formatting Completed")

# ============================
# STEP 7: VALIDATE NUMERIC DATA
# ============================

for col in numeric_cols:
    df[col] = pd.to_numeric(
        df[col],
        errors='coerce'
    )

print("\nNumeric Validation Completed")

# ============================
# STEP 8: FINAL QUALITY CHECK
# ============================

print("\nFINAL VALIDATION")

print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nDataset Shape After Cleaning:")
print(df.shape)

# ============================
# STEP 9: SAVE CLEAN DATASET
# ============================

df.to_csv(
    "cleaned_dataset.csv",
    index=False
)

print("\nCleaned Dataset Saved Successfully!")
```
